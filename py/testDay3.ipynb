max gap问题, 求数组最大差值，使用桶排序

```python
def max_gap(arr):
    gap = (max(arr) - min(arr))/(len(arr))
    min_arr = ['n']* (len(arr) + 1)
    max_arr = ['n']* (len(arr) + 1)
    for i in arr:
        num = int((i - min(arr))/gap)
        if min_arr[num] == 'n':
            min_arr[num] = i
        else:
            min_arr[num] = min(i,min_arr[num])
        if max_arr[num] == 'n':
            max_arr[num] = i
        else:
            max_arr[num] = max(i,max_arr[num])
    res = 0
    front = 0
    for i in range(1, len(min_arr)):
        if min_arr[i] == 'n':            
            continue
        res = max(min_arr[i] - max_arr[front], res)
        front = i
    return res
```

```python
arr = [-10,0,1,1,4,5,10,-21]
max_gap(arr)
```

固定数组实现栈和队列 pop push peek 

```python
class Stack:
    
    def __init__(self,size):
        self.size = size
        self.stack = [''] * size
        self.end = 0

    def isEmpty(self):
        if self.end == 0:
            return 1
        else:
            return 0

    def push(self,number):
        if self.end == self.size:
            print("stack is full")
        else:
            self.stack[self.end] = number 
            self.end += 1

    def peek(self):
        if Stack.isEmpty(self):
            print("stack is empty")
        else:
            tmp = self.end - 1
            return self.stack[tmp]
        
    def pop(self):
        if Stack.isEmpty(self):
            print("stack is empty")
        else:
            self.end -= 1
            return self.stack[self.end]

    def stackSize(self):
        return self.end
    
class Queue:
    
    def __init__(self,size):
        self.queue = [''] * size
        self.size = size
        self.sizeCurrent = 0
        self.end = 0
        self.start = 0
    
    def isEmpty(self):
        if self.sizeCurrent == 0:
            return 1
        else:
            return 0
    
    def push(self,number):
        if self.sizeCurrent == self.size :
            print("queue is full")
        else:
            self.queue[self.start] = number
            self.start += 1
            self.sizeCurrent += 1
            if(self.start >= self.size):
                self.start -= self.size
        
    def pop(self):
        if self.sizeCurrent == 0:
            print("queue is empty")
        else:
            temp = self.end 
            self.end  += 1
            self.sizeCurrent -= 1
            if(self.end >= self.size):
                self.end -= self.size
            return self.queue[temp]
    
    def peek(self):
        if self.sizeCurrent == 0:
            print("queue is empty")
        else:
            return self.queue[self.end]
        
    def queueSize(self):
        return self.sizeCurrent
```

```python
st = Stack(10)
print('this is a stack')
for i in range(0,5):
    print('push',i)
    st.push(i)
print('peek',st.peek())
print('size',st.stackSize())
for _ in range(0,3):
    print('pop',st.pop())
print('peek',st.peek()) 

qu = Queue(5)
print('this is a queue')
for i in range(0,5):
    print('push',i)
    qu.push(i)
for _ in range(0,3):
    print('pop',qu.pop())
for i in range(0,2):
    print('push',i)
    qu.push(i)
print('peek',qu.peek())
print('size',qu.queueSize())
for _ in range(0,5):
    print('pop',qu.pop())
```

用栈结构实现队列  //双栈 不断把备选栈往数据栈里倾倒 (当数据站为空时)
用队列实现栈    //双队列 把队列头N-1个放入另一个队列 吐出最后一个为pop

```python
class NewQueue:
    
    def __init__(self,size):
        self.size = size
        self.stack = Stack(size)
        self.data = Stack(size)
        
    def isEmpty(self):
        return self.data.isEmpty()
    
    def push(self,number):
        self.stack.push(number)
        self.swap()
    
    def pop(self):
        self.swap()
        return self.data.pop()
        
    def peek(self):
        return self.data.peek()
    
    def swap(self):
        if self.data.isEmpty():
            while(not self.stack.isEmpty()):
                self.data.push(self.stack.pop())
    
    def queueSize(self):
        return self.data.stackSize()
                
class NewStack:
    
    def __init__(self,size):
        self.size = size
        self.stack = Queue(size)
        self.data = Queue(size)
    
    def isEmpty(self):
        self.data.isEmpty()
    
    def push(self,number):
        self.data.push(number)
    
    def pop(self):
        while(self.data.queueSize() > 1):
            self.stack.push(self.data.pop())
        res = self.data.pop()
        self.swap()
        return res
    
    def peek(self):
        while(self.data.queueSize() > 1):
            self.stack.push(self.data.pop())
        res = self.data.pop()
        self.swap()
        self.data.push(res)
        return res
    
    def stackSize(self):
        return self.data.queueSize()
    
    def swap(self):
        temp = Queue(self.size)
        self.data = self.stack
        self.stack = temp
```

```python
st = NewStack(10)
print('this is a stack')
for i in range(0,5):
    print('push',i)
    st.push(i)
print('peek',st.peek())
print('size',st.stackSize())
for _ in range(0,3):
    print('pop',st.pop())
print('peek',st.peek()) 

qu = NewQueue(5)
print('this is a queue')
for i in range(0,5):
    print('push',i)
    qu.push(i)
for _ in range(0,3):
    print('pop',qu.pop())
for i in range(0,2):
    print('push',i)
    qu.push(i)
print('peek',qu.peek())
print('size',qu.queueSize())
for _ in range(0,5):
    print('pop',qu.pop())
```
